```python
crawlingThings_name_list=[]
crawlingThings_lot_list=[]
```

```python
from selenium import webdriver
import time

def crawling():
    
    time.sleep(1)
    crawlingThings = driver.find_elements_by_xpath("//li[@class='PlaceItem clickArea']")
    for crawlingThings in crawlingThings:
        crawlingThings_name = crawlingThings.find_element_by_class_name("link_name").text
        crawlingThings_lot = crawlingThings.find_element_by_class_name("lot_number").text.replace('(지번) ', '')
        crawlingThings_name_list.append(crawlingThings_name)
        crawlingThings_lot_list.append(crawlingThings_lot)
        print(crawlingThings_name, ':', crawlingThings_lot)
        
if __name__ == '__main__':
    
    driver = webdriver.Chrome(executable_path='chromedriver.exe')
    url="https://map.kakao.com/?nil_profile=title&nil_src=local"
    driver.get(url)
    
    time.sleep(1)
    
    searchBox = driver.find_element_by_id("search.keyword.query")
    searchBox.send_keys("광주 북구 유치원\n") #알고자 하는 장소 
    
    time.sleep(1)
    
    moreButton=driver.find_element_by_id("info.search.place.more")
    driver.execute_script("arguments[0].click();", moreButton)
    
pageNum = 1
while True:
    crawling()
    pageNum += 1
    if pageNum < 6:
        nextButton = driver.find_element_by_id(f"info.search.page.no{pageNum}")
        driver.execute_script("arguments[0].click();", nextButton)
    else:
        nextButton = driver.find_element_by_id(f"info.search.page.next")
        if nextButton.get_attribute('class') == 'next disabled':
            break
        else:
            pageNum = 1
            driver.execute_script("arguments[0].click();", nextButton)

print('\nProgram End')
```
